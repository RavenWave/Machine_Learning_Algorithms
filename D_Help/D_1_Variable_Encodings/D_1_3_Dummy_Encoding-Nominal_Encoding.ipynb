### Dummy Encoding

```python
import pandas as pd
```

```python
df0 = pd.read_csv("../../C_Datasets/titanic_train_original.csv")
df  = df0[['Sex', 'Embarked', 'Survived']]
df.head()
```

### 1) Default get_dummies()

```python
pd.get_dummies(df, columns=['Embarked'])
```

### 2) get_dummies() with prefix parameter

```python
pd.get_dummies(df, columns=['Embarked'], prefix='Emb', prefix_sep='-')
```

```python
pd.get_dummies(df, columns=['Embarked'], prefix='', prefix_sep='')
```

### 2) get_dummies() with drop_first parameter

```python
pd.get_dummies(df, columns=['Embarked'], drop_first=True)
```
