# **df.query()**

## **Import library and data set**

```python
import pandas as pd
```

```python
df = pd.read_csv('../../Machine_Learning_Algorithms/C_Datasets/Co2_Emission.csv').iloc[:, [1, 2, 3, 4, 5, 6, 7, 8, -1]]
df.head()
```

```python
df['FUELTYPE'].unique()
```

## **Filtering Rows Based on a Single Condition**

```python
df.query('ENGINESIZE > 8')
```

```python
df.query('CO2EMISSIONS < 120')
```

```python
df.query('CYLINDERS == 12')
```

```python
df.query("MAKE == 'ASTON MARTIN'")
```

```python
df.query("VEHICLECLASS in ('MINICOMPACT', 'SUBCOMPACT')")
```

## **Filtering Rows Based on Multiple Condition**

```python
df.query('ENGINESIZE > 4 and CO2EMISSIONS < 250')
```

```python
df.query('CYLINDERS == 6 and CO2EMISSIONS > 320')
```

```python
df.query("MAKE == 'BENTLEY' and CYLINDERS >= 8")
```

## **Handling Special Characters in Column Names**

If your column names contain `spaces` or `special characters`, you need to use backticks(`) around the column names.

```python
# convert FUELTYPE to FUEL TYPE
df.rename(columns={'FUELTYPE': 'FUEL TYPE'}, inplace=True)
```

```python
df.head(2)
```

```python
df.query("`FUEL TYPE` == 'X'")
```

```python
df.query("`FUEL TYPE` in ('X', 'Y')")
```

```python
df.query("`FUEL TYPE` in ('X', 'Y') and MAKE == 'VOLVO'")
```

## **Using Variables in the Query**

```python
avg_emission = df['CO2EMISSIONS'].mean()

print('Average Co2 Emission is {}.\n'.format(round(avg_emission)))

df.query('CO2EMISSIONS > @avg_emission')
```

```python
max_engine_size_volvo = df.query("MAKE == 'VOLVO'")['ENGINESIZE'].max()

df.query("MAKE == 'VOLVO' and ENGINESIZE == @max_engine_size_volvo")
```

## **More Complex Queries**

You can use methods and complex expressions within the `.query()` function. 

```python
df.query("MAKE.str.startswith('V')")
```

```python
df.query("MAKE.str.startswith('V') and MODEL.str.startswith('S')")
```

```python
df.query("VEHICLECLASS.str.contains('SUV')")
```

```python
df.query("VEHICLECLASS.str.lower().str.contains('suv')")
```

```python
df.query('MODEL.notna()') # notnull()
```

```python
df.query('MODEL.isna()') # or --> isnull()
```

```python
df.query('3 <= CYLINDERS <= 8')
```

```python
df.query("`FUEL TYPE` == ('X', 'Y')") 
```

```python
df.query("ENGINESIZE.astype('str').str.startswith('3')")
```

```python
df.query("VEHICLECLASS.str.endswith('E', na=False)") # if there is null values, you can use parameter of 'na=False'
```

```python
df.query('index % 2 == 0')
```

```python
df.query("not MAKE == 'VOLVO'")
```

# **df.filter()**

## **Creating a Data Frame**

```python
df = pd.DataFrame([[0, 1, 2], [3, 4, 5], [6, 7, 8]],
                  index=['apple', 'banana', 'pineapple'],
                  columns=['A', 'B', 'C'])
```

```python
df
```

## **Basic Usage**

```python
df.filter(items=['A', 'C'])
```

```python
df.filter(items=['A', 'C'], axis=None)   # default value of axis parameter is 'None' 
```

```python
df.filter(items=['A', 'C'], axis=1)   # axis=1 and axis=None are equal each other
```

```python
df.filter(items=['A', 'C'], axis='columns')   # axis=1 and axis='columns' are equal each other
```

```python
df.filter(items=['apple', 'banana'], axis=0)   # axis=1 filters based on rows.
```

```python
df.filter(items=['apple', 'banana'], axis='index')   # axis=0 and axis='index' are equal each other
```

```python
# to filter columns and rows together, we must repeat the filter() function.

df.filter(items=['A', 'C']).filter(items=['pineapple'], axis=0)
```

## **the 'like' parameter**

```python
df = pd.DataFrame([[0, 1, 2], [3, 4, 5], [6, 7, 8]],
                  index=['apple', 'banana', 'pineapple'],
                  columns=['col X', 'col Y', 'col Z'])
```

```python
df
```

```python
df.filter(like='Y')
```

```python
df.filter(like='apple', axis=0)
```

```python
# combine axis 0 and axis 1

df.filter(like='Y').filter(like='apple', axis=0)
```

# **Extract rows that contain specific strings**

## **Exact match with spesific string: `==`, `isin()`**

```python
df[df.MODEL == 'ILX']
```

```python
df[df.VEHICLECLASS == 'COMPACT']
```

```python
df[df.FUELTYPE.isin(['D', 'X'])]
```

```python
df[df.VEHICLECLASS.isin(['MID-SIZE', 'SUV - SMALL'])]
```

## **Partial match (Contains specific string): `str.contains()`**

```python
df[df['MODEL'].str.contains('WD')]
```

```python
df[df['VEHICLECLASS'].str.contains('SUV')]
```

```python
# Using the parameter 'na' as True, we could include the Null values. (default value of parameter na is False),
# We don't have any null values in our data frame, which we don't see any null values in the output.

df[df['VEHICLECLASS'].str.contains('SUV', na=True)]
```

```python
# By default, `str.contains()` distinguishes between uppercase and lowercase characters. To ignore case distinction, set the `case` argument to `False`.

df[df['VEHICLECLASS'].str.contains('SUV', case=False)]
```

## **Forward match (Starts with a specific string): `str.startswith()`**

```python
df[df['MODEL'].str.startswith('XC')]
```

## **Backward match (Ends with a specific string): `str.endswith()`**

```python
df[df['MODEL'].str.endswith('4X4')]
```

## **String starts with a match of a regular expression: `str.match()`**

```python
df[df['VEHICLECLASS'].str.match('^PICKUP')]
```

```python
df[df['VEHICLECLASS'].str.match('^PICKUP.*SMALL$')]
```
